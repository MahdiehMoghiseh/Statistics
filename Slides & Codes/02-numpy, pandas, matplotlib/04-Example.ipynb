```python
import numpy as np

# people = []
# for i in range(50):
#   people.append(100)

# people = [100 for i in range(50)]

people = np.full(50, 100)
# len(people)
people
```

```python
import random

random.seed()
for j in range(10000):

  for i in range(0, 50):
    receive = random.randint(0,49)
    while people[receive] == 0:
      receive = random.randint(0,49)
    if people[i] != 0:
      people[i] -= 1
      people[receive] += 1
  
people
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize= (15 , 15))
people.sort()
plt.bar(range(0,50) , people)
```
